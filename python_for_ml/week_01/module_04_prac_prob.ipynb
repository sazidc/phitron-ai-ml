```python
brightness, threshold = map(float, input().split())

if brightness >= threshold:
    print("ON")
else:
    print("OFF")
```

```python
x, min_v, max_v = map(float, input().split())

norm = (x - min_v) / (max_v - min_v)

print(f"{norm:.2f}")
```

```python
N = int(input())
target = float(input())

total = 0.0

for _ in range(N):
    loss = float(input())
    total += loss

average = total / N

if average <= target:
    print("PASS")
else:
    print("RETRY")
```

```python
N = int(input())

yes_count = 0
no_count = 0

for _ in range(N):
    vote = input().strip()
    if vote == "YES":
        yes_count += 1
    else:
        no_count += 1

if yes_count >= no_count:
    print("ACCEPT")
else:
    print("REJECT")
```

```python
message = input().strip()

if "happy" in message or "joy" in message or "smile" in message:
    print("Happy Mood")
elif "sad" in message or "cry" in message or "angry" in message:
    print("Sad Mood")
else:
    print("Neutral Mood")
```

```python
# Read the input list of integers
pixels = list(map(int, input().split()))

# Compute the average brightness
average_brightness = sum(pixels) / len(pixels)

# Determine the image type based on average brightness
if average_brightness < 85:
    print("Dark Image")
elif 85 <= average_brightness <= 170:
    print("Normal Image")
else:
    print("Bright Image")
```

```python
predictions = input().split()

total = len(predictions)
count_A = 0
count_B = 0

# Count A and B
for p in predictions:
    if p == "A":
        count_A += 1
    else:
        count_B += 1

# Check if either exceeds 70%
if count_A > 0.7 * total or count_B > 0.7 * total:
    print("Biased Model")
else:
    print("Fair Model")
```

```python
message = input().strip()

keywords = ["ai", "data", "model", "learn", "train", "neural"]

count = 0

for k in keywords:
    if k in message:
        count += 1

if count >= 2:
    print("AI Detected")
else:
    print("Not AI Related")
```
